<font color=blue>**Paso 2**</font> Definición y operaciones entre matrices

Definir las matrices $A=\left [ \begin{matrix} 1 & 2 & 3\\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{matrix} \right ]$ y $B=\left [ \begin{matrix} 2 & 0 & 0\\ 0 & 2 &  0 \\ 0 & 0 & 2 \end{matrix} \right ]$, y tenemos dos maneras

```r
A = matrix(1:9, nrow=3, ncol=3, byrow=TRUE)
print(A)
```

```r
A=rbind(1:3, 4:6, 7:9)
print(A)
```

```r
A=cbind(c(1,4,7),c(2,5,8),c(3,6,9))
print(A)
```

```r
B=2*diag(3)
print(B)
```

Todas las **operaciones básica** con vectores y matrices (+,-,\*) son por default *elemento-a-elemento*

```r
print(A+B)
print(A*B)
print(A%*%B)
```

<font color=blue>**Paso 6**</font> Elementos básicos de graficación

Para graficar se emplean **vectores** (*longitud igual*) para los ejes-*x* y eje-*y*

```r
N=1000
x=seq(0,10,length.out=N)
y=exp(-x)*sin(2*pi*x)
plot(x,y,type='l',xlab='x', ylab='f(x)', main='Grafica de una Sinusoide Amortiguada', col=2, lwd=2)
grid()
```

```r
N=length(y)
n=rnorm(N,mean=0, sd=0.05)
plot(x,y+n,type='l',xlab='x', ylab='f(x)', main='Grafica de una Sinusoide Amortiguada+ Ruido', col=3, lwd=2, lty=2)
grid()
```

Las gráficas se pueden **imprimir** como archivos con extensión de imagen --> PNG, JPEG, PS, PDF, SVG, BMP, TIFF

```r
png('plot1.png')
plot(x,y+n,type='l',xlab='x', ylab='f(x)', main='Grafica de una Sinusoide Amortiguada', col=3, lwd=2, lty=2)
grid()
dev.off()
```

```r
jpeg('plot1.jpg')
plot(x,y+n,type='l',xlab='x', ylab='f(x)', main='Grafica de una Sinusoide Amortiguada', col=3, lwd=2, lty=2)
grid()
dev.off()
```
